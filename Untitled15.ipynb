```python
eleccion=0
func=0
cv=150
cs=180
cvsold=0
cssold=0
cafel=[]
def comprar_entradas():
    global cs, cv, cssold, cvsold, func
    while True:
        nombre=str(input("""-- Comprar entrada a Cats --
        Nombre del comprador: """)).lower().strip()
        if nombre in cafel:
            print("el nombre ya esta registrado")
            break
        else:
            print(f"""Seleccione función:
        1. Cats Día Viernes (150 entradas)
        2. Cats Día Sábado (180 entradas)""")
            func=int(input("Función (1 o 2): "))
            if func==1:
                cv=cv-1
                cvsold=cvsold+1
                cafel.append({
                    "nombre" : nombre,
                    "func" : func
                    })
                print(f"""Entrada registrada en función {func}! Stock restantes:
        Función 1 (Viernes): {cv}
        Función 2 (Sábado): {cs}""")
                break
            elif func ==2:
                cs=cs-1
                cssold=cssold+1
                cafel.append({
                    "nombre" : nombre,
                    "func" : func
                    })
                print(f"""Entrada registrada en función {func}! Stock restantes:
        Función 1 (Viernes): {cv}
        Función 2 (Sábado): {cs}""")
                break
            else:
                print("Error: opción de función inválida")
                break


def cambio_funcion():
    global cafel, cv, cs, cvsold, cssold
    nombrebus = input("""-- Cambio de función --
Nombre del comprador: """).strip().lower()
    for comprador in cafel:
        if comprador["nombre"] == nombrebus:
            actual = comprador["func"]
            if actual == 1:
                if cs <= 0:
                    print("No hay stock disponible para la función del sábado.")
                    return
                confirm = input("Actualmente tiene entrada para el viernes. ¿Desea cambiar al sábado? (S/N): ").strip().lower()
                if confirm == "s":
                    comprador["func"] = 2
                    cv += 1
                    cs -= 1
                    cvsold -= 1
                    cssold += 1
                    print("¡Cambio realizado con éxito!")
                else:
                    print("Cambio cancelado.")
            elif actual == 2:
                if cv <= 0:
                    print("No hay stock disponible para la función del viernes.")
                    return
                confirm = input("Actualmente tiene entrada para el sábado. ¿Desea cambiar al viernes? (S/N): ").strip().lower()
                if confirm == "s":
                    comprador["func"] = 1
                    cs += 1
                    cv -= 1
                    cssold -= 1
                    cvsold += 1
                    print("¡Cambio realizado con éxito!")
                else:
                    print("Cambio cancelado.")
            return
    print("Nombre no encontrado.")

def mostrar_totales():
    global cs, cv, cssold, cvsold
    print(f"""-- Stock de Funciones --
Función 1 (Viernes): Disponibles {cv}, Vendidas {cvsold}
Función 2 (Sábado): Disponibles {cs}, Vendidas {cssold}
""")
    input("Presione enter para continuar")

def menu():
    eleccion=0
    while eleccion!=4:
        print("")
        print("""TOTEM AUTOATENCIÓN CAFECONLECHE
1.- Comprar entrada a Cats.
2.- Cambio de función.
3.- Mostrar stock de funciones.
4.- Salir
        """)
        eleccion=int(input("Eliga la opcion deseada: "))
        if eleccion==1:
            comprar_entradas()
        elif eleccion ==2:
            cambio_funcion()
        elif eleccion ==3:
            mostrar_totales()
        elif eleccion == 4:
            print("Programa terminado...")
        elif eleccion == 5:
            print(cafel)
        else:
            print("Debe ingresar una opción válida!!")


menu()
```
